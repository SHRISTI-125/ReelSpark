```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
texts = []
metadatas = []
```

```python
import json
```

```python
print(type(db))
print(type(next(iter(db))))
```

```python
for _, doc in db.iterrows():
    combined_text = f"""
    Task: {doc['task']}
    Platform: {doc['platform']}
    Content Type: {doc['content_type']}
    Intent: {doc['intent']}
    Tone: {doc['tone']}
    Structure: {doc['structure']}
    Context: {doc['input_context']}
    Output Example: {doc['output_text']}
    """
    texts.append(combined_text.strip())
    metadatas.append(doc.to_dict())
```

```python
print(f"Prepared {len(texts)} of documents")
```

```python
embeddings = model.encode(
    texts,
    convert_to_numpy=True,
    normalize_embeddings=True
)
```

```python
print("Embeddings shape:", embeddings.shape)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
model_name = "google/flan-t5-small"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name, local_files_only=True)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name, local_files_only=True)
```

```python
def generate_variation(text, num_variations=1):
    prompt = f"paraphrase: {text}"
    inputs = tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True)
    outputs = model.generate(
        **inputs,
        max_length=64,
        num_return_sequences=num_variations,
        do_sample=True,
        top_k=50,
        top_p=0.95
    )
    return [tokenizer.decode(o, skip_special_tokens=True) for o in outputs]
```
